```python
import pandas as pd
import matplotlib.pyplot as plt

# Creating a DataFrame with your data
data = {
    'Item': ['God', 'Angels', 'Heaven', 'Hell', 'The devil'],
    '% Believe in': [74, 69, 67, 59, 58],
    '% Not sure about': [14, 13, 15, 14, 14],
    '% Don’t believe in': [12, 18, 18, 27, 28]
}

df = pd.DataFrame(data)

# Plotting the data
df.set_index('Item').plot(kind='bar', figsize=(10,6))
plt.title('Belief in Various Concepts')
plt.xlabel('Concepts')
plt.ylabel('Percentage')
plt.xticks(rotation=0)
plt.legend(title='Belief Status')
plt.show()
```
