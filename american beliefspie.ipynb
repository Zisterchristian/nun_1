```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Creating the DataFrame
data = {
    'Item': ['God', 'Angels', 'Heaven', 'Hell', 'The devil'],
    '% Believe in': [74, 69, 67, 59, 58],
    '% Not sure about': [14, 13, 15, 14, 14],
    '% Don’t believe in': [12, 18, 18, 27, 28]
}

df = pd.DataFrame(data)

# Setting the index to 'Item'
df.set_index('Item', inplace=True)

# Creating the heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(df, annot=True, cmap='coolwarm', fmt='d')
plt.title('Belief in Various Concepts')
plt.xlabel('Belief Status')
plt.ylabel('Concepts')
plt.show()
```
