```python
import pandas as pd
import matplotlib.pyplot as plt

# Creating a DataFrame with your data
data = {
    'Item': ['God', 'Angels', 'Heaven', 'Hell', 'The devil'],
    '% Believe in': [74, 69, 67, 59, 58],
    '% Not sure about': [14, 13, 15, 14, 14],
    '% Dont believe in': [12, 18, 18, 27, 28]
}

df = pd.DataFrame(data)

# Plotting the pie chart for '% Believe in'
plt.figure(figsize=(8, 8))
plt.pie(df['% Believe in'], labels=df['Item'], autopct='%1.1f%%', startangle=140)
plt.title('% Believe in Various Concepts')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Creating a DataFrame with your data
data = {
    'Item': ['God', 'Angels', 'Heaven', 'Hell', 'The devil'],
    '% Believe in': [74, 69, 67, 59, 58],
    '% Not sure about': [14, 13, 15, 14, 14],
    '% Dont believe in': [12, 18, 18, 27, 28]
}

df = pd.DataFrame(data)

# Plotting the scatter chart
plt.figure(figsize=(10, 6))
for i, row in df.iterrows():
    plt.scatter(['% Believe in', '% Not sure about', '% Dont believe in'], 
                [row['% Believe in'], row['% Not sure about'], row['% Dont believe in']],
                label=row['Item'])

plt.title('Belief in Various Concepts')
plt.xlabel('Belief Status')
plt.ylabel('Percentage')
plt.legend(title='Concepts')
plt.grid(True)
plt.show()
```
